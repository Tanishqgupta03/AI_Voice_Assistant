So the application is going to look something like this the voice will be the input and then gives 2 things as the output one is text and that text in voice.

```python
def voice_chat(user_voice):

  messages = [{
      "role": "system", "content": "You are a Kind helpful assistant."
  }]#First setting up the helpful assistant so that gpt knows that it needs to help us

  user_message = model.transcribe(user_voice)["text"]#u take user voice trancribe it and assign to a object

  messages.append(
      {"role":"user", "content":user_message},
  )# now send the message into the messages list where we already have set the behaviour

  print(messages)

  chat = openai.ChatCompletion.create(
      model="gpt-3.5-turbo", messages=messages
  )#api call

  reply = chat.choices[0].message.content
  messages.append({"role": "assistant", "content":reply})# append the reply
  # as agr chat phele se chal rahi hai then gpt could be able to connect with the
  # the previous chats

  tts.tts_to_file(text=reply, file_path="output.wav")

  return(reply, 'output.wav')
```

Building the gradio application, can be made using 2 methods 1 is the interface method 2nd is the blocks methid.

using interface method.
It needs 3 things input output and the function itself

```python
text_reply = gr.Textbox(label="ChatGPT Text")
voice_reply = gr.Audio('output.wav')

gr.Interface(
    title = 'AI Voice Assistant With ChatGPT AI',
    fn=voice_chat,
    inputs=[
        gr.inputs.Audio(source="microphone", type="filepath")
    ],

    outputs=[
        text_reply, voice_reply
    ], live = True).launch(debug = True)
```
